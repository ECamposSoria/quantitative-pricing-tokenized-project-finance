# 02 · Excel vs Python
Compara los resultados exportados en `TP_Quant_Validation.xlsx` con el pipeline.

```python
import pandas as pd
import openpyxl
from pftoken.models import ProjectParameters
from pftoken.pipeline import FinancialPipeline

params = ProjectParameters.from_directory('data/input/leo_iot')
pipeline = FinancialPipeline(params=params)
results = pipeline.run()
excel = openpyxl.load_workbook('TP_Quant_Validation.xlsx', data_only=True)
cfads_python = results['cfads']
cfads_excel = {row[0].value: row[8].value for row in excel['CFADS_Calc'].iter_rows(min_row=2, max_row=16)}
comparison = pd.DataFrame({
    'Year': list(cfads_python.keys()),
    'Python': list(cfads_python.values()),
    'Excel': [cfads_excel[year] for year in cfads_python.keys()],
})
comparison
```
