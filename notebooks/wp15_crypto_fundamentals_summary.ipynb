# WP-15 Crypto Fundamentals (Scoped)

Minimal deliverable: document chain/platform choice, add regulatory tail-risk premium, and keep non-pricing tasks out of scope.

## Chain Finality & Costs
- Centrifuge/Polkadot: ~6s deterministic finality; low fees suitable for annual CFADS + quarterly reporting cadence.
- Ethereum L1: ~$5–$20/tx, probabilistic finality; too costly for routine reporting.
- Arbitrum (reference L2): ~$0.10–$0.50/tx; finality faster than reporting cadence → no finality premium required.
- Decision: use Centrifuge/Tinlake rails; no finality premium priced.

## Platform Economics
- **Centrifuge/Tinlake**: ~$1.45B TVL, ~15 bps protocol fee, CFG incentives, no sponsor LP capital required.
- **Maple Finance**: institutional focus, higher minimums and fees than 15 bps protocol rate.
- **Goldfinch**: higher LP spreads, concentrated manager risk.
- **Traditional SPV servicer**: 50–100 bps servicing/admin with no protocol incentives.
- Selection rationale: choose Centrifuge for lowest fee + incentives + RWA-native infra.

## Regulatory Risk
- Tail-risk scenario: security token regulatory ban (T3 stress).
- Premium applied: 7.5 bps (midpoint of 5–10 bps range) in `wacd_synthesis`.
- Tokenization benefits remain (liquidity/operational/transparency); premium is subtracted to reflect tail risk.
